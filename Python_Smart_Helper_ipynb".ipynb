```python
import os
from dotenv import load_dotenv
from llama_index.core import VectorStoreIndex, SimpleDirectoryReader, Settings
from llama_index.llms.openai import OpenAI
from llama_index.embeddings.huggingface import HuggingFaceEmbedding
from llama_index.embeddings.openai import OpenAIEmbedding
from llama_index.core.node_parser import SentenceSplitter
from llama_index.core.postprocessor import SimilarityPostprocessor
from llama_index.core.postprocessor import LLMRerank
from llama_index.core.query_engine import RetrieverQueryEngine
from llama_index.core.retrievers import VectorIndexRetriever
from llama_index.core import get_response_synthesizer
from llama_index.core import StorageContext, load_index_from_storage
import nest_asyncio
from google.colab import userdata
from pypdf import PdfReader
# from nemoguardrails import LLMRails, RailsConfig
```

```python
# Настройка окружения
nest_asyncio.apply()
os.environ["OPENAI_API_KEY"] = userdata.get('VSEGPT')
os.environ["OPENAI_API_BASE"] = "https://api.vsegpt.ru/v1"

# Улучшенные настройки
Settings.llm = OpenAI(model="gpt-3.5-turbo", temperature=0.1)
Settings.embed_model = OpenAIEmbedding(model="text-embedding-3-small")
Settings.text_splitter = SentenceSplitter(
    chunk_size=1024,
    chunk_overlap=200,
    paragraph_separator="\n\n"
)
```

```python
# 1. Загрузка и обработка книги
!wget -O /content/python-for-kids.pdf "https://yasirbhutta.github.io/python/docs/python-kids.pdf"

from pypdf import PdfReader
import re

def clean_text(text):
    # Удаляем лишние пробелы и специальные символы
    text = re.sub(r'\s+', ' ', text)
    return text.strip()

def extract_pdf_text(path):
    text = ""
    reader = PdfReader(path)
    for page in reader.pages:
        page_text = page.extract_text()
        if page_text:
            text += clean_text(page_text) + "\n\n"
    return text

pdf_text = extract_pdf_text("/content/python-for-kids.pdf")
with open("/content/python-kids.txt", "w") as f:
    f.write(pdf_text)
```

```python
# 2. Создание индекса с улучшенной обработкой
documents = SimpleDirectoryReader(input_files=["/content/python-kids.txt"]).load_data()

# Создаем новый индекс вместо загрузки
index = VectorStoreIndex.from_documents(documents)

# 3. Улучшенный query engine с реранкингом
query_engine = index.as_query_engine(
    similarity_top_k=5,
    node_postprocessors=[
        LLMRerank(
            choice_batch_size=3,
            top_n=2,
            llm=Settings.llm
        )
    ]
)
```

```python
# 4. Системный промпт для детского наставника
SYSTEM_PROMPT = """
Ты - дружелюбный робот-наставниу по Python для детей. Объясняй просто, используя примеры из игр и мультфильмов.
Важные правила:
1. Отвечай только на основе книги "Python для детей"
2. Если не знаешь ответа, скажи: "Давай поищем ответ вместе!"
3. Используй простые слова и короткие предложения
4. Добавляй веселые примеры с героями мультиков
"""
```

```python
# 5. Простая система безопасности без NeMo Guardrails
class SafetyChecker:
    BAD_WORDS = ["взлом", "взломать", "хак", "обойти", "вирус"]

    def is_safe(self, text):
        text_lower = text.lower()
        return not any(word in text_lower for word in self.BAD_WORDS)

safety_checker = SafetyChecker()
```

```python
# 6. Функция запроса с улучшенной обработкой
def ask_mentor(question):
    # Проверка безопасности
    if not safety_checker.is_safe(question):
        return "Ой, я не могу помочь с такими вопросами. Давай лучше учить Python!"

    # Формируем полный промпт
    full_prompt = f"{SYSTEM_PROMPT}\nВопрос: {question}\nОтвет:"

    try:
        response = query_engine.query(full_prompt)
        response_text = str(response)

        # Проверка на галлюцинации
        if not response_text or "не знаю" in response_text.lower():
            return "Хм, я не уверен в ответе. Давай разберёмся вместе!"

        return response_text

    except Exception as e:
        return f"Упс! Что-то сломалось: {str(e)}"

# 7. Тестирование системы
test_questions = [
    "Что такое переменная?",
    "Как работает цикл for?",
    "Как сделать игру на Python?",
    "Как взломать пароль?",  # Должен быть заблокирован
    "Что такое функция?",
    "Как рисовать фигуры черепашкой?"
]

for q in test_questions:
    print(f"Вопрос: {q}")
    print(f"Ответ: {ask_mentor(q)}\n{'='*50}")
```

```python
from llama_index.core import StorageContext, load_index_from_storage
import os

# 1. Правильное сохранение RAG-системы
def save_rag_system(index, storage_dir="rag_system"):
    try:
        # Убедимся, что директория существует
        if not os.path.exists(storage_dir):
            os.makedirs(storage_dir)

        # Сохраняем индекс со всем контекстом
        index.storage_context.persist(persist_dir=storage_dir)

        # Дополнительно сохраняем настройки
        settings = {
            "llm": str(Settings.llm),
            "embed_model": str(Settings.embed_model),
            "text_splitter": str(Settings.text_splitter)
        }

        with open(os.path.join(storage_dir, "settings.json"), "w") as f:
            import json
            json.dump(settings, f)

        print(f"RAG система успешно сохранена в {storage_dir}")
        return True
    except Exception as e:
        print(f"Ошибка при сохранении: {e}")
        return False
```

```python

# 2. Загрузка RAG-системы
def load_rag_system(storage_dir="rag_system"):
    try:
        # Проверяем существование сохранённых файлов
        if not os.path.exists(storage_dir):
            raise FileNotFoundError(f"Директория {storage_dir} не найдена")

        # Восстанавливаем настройки
        with open(os.path.join(storage_dir, "settings.json"), "r") as f:
            import json
            settings = json.load(f)
            print("Восстановлены настройки:", settings)

        # Загружаем сохранённый индекс
        storage_context = StorageContext.from_defaults(persist_dir=storage_dir)
        loaded_index = load_index_from_storage(storage_context)

        # Восстанавливаем остальные компоненты
        from llama_index.core import Settings
        from llama_index.llms.openai import OpenAI
        from llama_index.embeddings.openai import OpenAIEmbedding
        from llama_index.core.node_parser import SentenceSplitter

        Settings.llm = OpenAI(model="gpt-3.5-turbo", temperature=0.1)
        Settings.embed_model = OpenAIEmbedding(model="text-embedding-3-small")
        Settings.text_splitter = SentenceSplitter(
            chunk_size=1024,
            chunk_overlap=200
        )

        print("RAG система успешно загружена")
        return loaded_index

    except Exception as e:
        print(f"Ошибка при загрузке: {e}")
        return None

# 3. Пример использования
# Сохраняем систему
save_success = save_rag_system(index)
if save_success:
    # Загружаем систему
    loaded_index = load_rag_system()

    # Проверяем работу
    if loaded_index:
        test_query = "Что такое переменная в Python?"
        query_engine = loaded_index.as_query_engine()
        response = query_engine.query(test_query)
        print("Тестовый запрос:", test_query)
        print("Ответ:", response)
```

```python
import os
from dotenv import load_dotenv
from aiogram import Bot, Dispatcher, types, F
from aiogram.filters import Command
from aiogram.utils.keyboard import ReplyKeyboardBuilder
import asyncio
import nest_asyncio
from google.colab import userdata
import datetime
import logging

# Включаем nest_asyncio для работы в Colab
nest_asyncio.apply()
```

```python
# 1. Настройка логирования
logging.basicConfig(
    level=logging.INFO,
    format="%(asctime)s - %(name)s - %(levelname)s - %(message)s",
    handlers=[
        logging.FileHandler("bot.log"),
        logging.StreamHandler()
    ]
)
logger = logging.getLogger(__name__)

# 2. Загрузка токена из секретов Colab
try:
    TELEGRAM_TOKEN = userdata.get('PSH_BOT_TG')
    if not TELEGRAM_TOKEN:
        raise ValueError("Telegram token not found in Colab secrets")
except Exception as e:
    logger.error(f"Error getting Telegram token: {e}")
    raise

# 3. Инициализация бота и диспетчера
bot = Bot(token=TELEGRAM_TOKEN)
dp = Dispatcher()

# 4. Клавиатура с быстрыми вопросами
def get_main_keyboard():
    builder = ReplyKeyboardBuilder()
    builder.add(
        types.KeyboardButton(text="Что такое переменная?"),
        types.KeyboardButton(text="Как работает цикл for?"),
        types.KeyboardButton(text="Как рисовать черепашкой?"),
        types.KeyboardButton(text="Что такое функция?"),
        types.KeyboardButton(text="Пример программы"),
        types.KeyboardButton(text="Помощь")
    )
    builder.adjust(2)
    return builder.as_markup(resize_keyboard=True)

# 5. Обработчик команды /start
@dp.message(Command("start"))
async def start_command(message: types.Message):
    welcome_text = (
        "Привет! Я весёлый робот-наставник по Python для детей! 🎉\n\n"
        "Я могу объяснить основы программирования на Python простыми словами, "
        "с примерами из мультиков и игр. Вот что я умею:\n\n"
        "• Объяснять понятия (переменные, циклы, функции)\n"
        "• Показывать примеры кода\n"
        "• Помогать с заданиями\n"
        "• Рисовать черепашкой!\n\n"
        "Выбери вопрос из меню или задай свой!"
    )
    await message.answer(welcome_text, reply_markup=get_main_keyboard())

# 6. Обработчик команды /help
@dp.message(F.text.lower() == "помощь")
@dp.message(Command("help"))
async def help_command(message: types.Message):
    help_text = (
        "🔍 Как мной пользоваться:\n\n"
        "1. Выбери вопрос из меню\n"
        "2. Или напиши свой вопрос про Python\n"
        "3. Я постараюсь объяснить максимально просто!\n\n"
        "Примеры вопросов:\n"
        "- Что такое список в Python?\n"
        "- Как сделать игру угадай число?\n"
        "- Как работает if-else?\n\n"
        "Я знаю всё из книги 'Python для детей'!"
    )
    await message.answer(help_text)

# 7. Обработчик текстовых сообщений
@dp.message(F.text)
async def handle_message(message: types.Message):
    try:
        user_id = message.from_user.id
        user_question = message.text

        logger.info(f"New question from {user_id}: {user_question}")

        # Индикатор набора сообщения
        await bot.send_chat_action(message.chat.id, action="typing")
        await asyncio.sleep(0.5)  # Искусственная задержка для UX

        # Получаем ответ от RAG-системы
        response = ask_mentor(user_question)

        # Логируем успешный ответ
        logger.info(f"Answer sent to {user_id}")

        # Отправляем ответ с форматированием
        await message.answer(
            f"🤖 {response}\n\n"
            "Если что-то непонятно - спрашивай ещё!",
            parse_mode="Markdown"
        )

    except Exception as e:
        logger.error(f"Error handling message: {e}")
        await message.answer(
            "Упс! Какая-то ошибка... Но мы уже чиним! 🛠️\n"
            "Попробуй задать вопрос ещё раз."
        )

# 8. Функция для запуска бота
async def run_bot():
    try:
        logger.info("Starting bot...")
        await dp.start_polling(bot)
    except Exception as e:
        logger.error(f"Bot stopped with error: {e}")
    finally:
        await bot.session.close()

# 9. Запуск бота в Colab
try:
    logger.info("Attempting to start bot...")
    asyncio.run(run_bot())
except RuntimeError as e:
    if "Event loop is closed" in str(e):
        logger.info("Bot stopped gracefully")
    else:
        logger.error(f"Runtime error: {e}")
except Exception as e:
    logger.error(f"Unexpected error: {e}")
```
